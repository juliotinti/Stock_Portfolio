<h1> Creation of used classes </h1>

```python
class FinancialInput:
    def __init__(self, date, totalValue, quantity):
        self.date = date
        self.totalValue = totalValue
        self.quantity = quantity
        self.pricePerAsset = self.totalValue/self.quantity

class FinancialAssets:
    def __init__(self, name, ticker, entry, sector, idealPercentage, targetPrice, intrinsicValue): 
        self.name = name
        self.ticker = ticker
        self.entry = entry
        self.sector = sector
        self.myDividend = []
        self.financialInput = []
        self.idealPercentage = idealPercentage
        self.targetPrice = targetPrice
        self.intrinsicValue = intrinsicValue
        
    def getName(self):
        return self.name
    
    def getSector(self):
        return self.sector
    
    def getMyDividend(self): 
        return self.myDividend 
    def addMyDividend(self, dividend): 
        self.myDividend.append(dividend)
        
    def getCurrentPrice(self):
        return self.ticker.info["currentPrice"]
        
    def getInvestedValue(self):
        return sum(self.financialInput.totalValue)
        
    def getIdealPercentage(self):
        return self.idealPercentage
    
    def getCurrentPrice(self):
        return self.ticker.info["currentPrice"]
    
    def getQuantity(self):
        return sum(self.financialInput.quantity)
    
    def getMyCurrentAmount(self):
        return (self.getCurrentPrice()*sum(self.getQuantity()))
    
    def getTargetPrice(self):
        return self.targetPrice
    def setTargetPrice(self, target):
        self.targetPrice = target
    
    def getAvgPrice(self):
        return sum(self.getInvestedValue()/self.getQuantity())
        
    def getIntrinsicValue (self):
        return self.intrinsicValue 
    def setIntrinsicValue (self, value):
        self.intrinsicValue = value   
        
    def getEntry(self):
        return self.entry
    
    def myDividendYield(self):
        return sum(self.myDividend)/self.getMyCurrentAmount()
    
    def myYieldOnCost(self):
        return sum(self.myDividend)/self.getInvestedValue()
    
    def balanceSheet(self):
        return self.ticker.balance_sheet
    
    def financialsInfo(self):
        return self.ticker.financials
    
    def variation(self):
        return ((self.getCurrentPrice()/self.getAvgPrice())-1)
    
class BrlStocks (FinancialAssets):
    def __init__(self):
        self.financialType = "Ações"
        self.walletPercentage = 0.24      
    
class AmrStocks (FinancialAssets):
    def __init__(self):
        self.financialType = "Stocks"
        self.walletPercentage = 0.24
        
class Fiis (FinancialAssets): 
    def __init__(self):
        self.financialType = "Fundo Imobiliário"
        self.walletPercentage = 0.24
        
class Reits (FinancialAssets):
    def __init__(self):
        self.financialType = "Reits"
        self.walletPercentage = 0.24

#TEM QUE MUDAR ALGUMAS COISA, PREÇO MÉDIO, e outras paradas...PENSAR DPS
class Cryptos (): 
    def __init__(self, ticker, quantity, entry, idealPercentage, info):
        self.financialType = "Crypto"
        self.walletPercentage = 0.04 
        self.ticker = ticker
        self.currentPrice = self.info["currentPrice"]
        self.quantity = quantity
        self.entry = entry
        self.idealPercentage = idealPercentage
        self.info = info
        self.investedValue = []
        self.avgPrice = avgPrice
```
